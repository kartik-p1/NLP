```python


def extract_second_elements(list_of_tuples):
    return [tup[1] for tup in list_of_tuples]

# Example usage:
list_of_tuples = [('अतिया', 'B-PER'), ('ने', 'O'), ('बताया', 'O')]
result_array = extract_second_elements(list_of_tuples)
file_path = "BertOutput.txt"
BertOutput = []
with open(file_path, "r", encoding="utf-8") as file:
    for line in file:
        line = eval(line.strip())  # Assumes the lines are in the format of Python tuple literals
        result_array = extract_second_elements(line)
        for i in result_array:
            BertOutput.append(i)
        
# print("                          -------------------------                                        ")
file_path = "ground_truth_for_bert.txt"
ground_truth = []
with open(file_path, "r", encoding="utf-8") as file:
    for line in file:
        line = eval(line.strip())  # Assumes the lines are in the format of Python tuple literals
        result_array = extract_second_elements(line)
        for i in result_array:
            ground_truth.append(i)
```

```python
tags = ['O', 'B-PER', 'I-PER', 'B-LOC', 'I-LOC', 'B-ORG', 'I-ORG', 'B-MISC', 'I-MISC']
def calculate_metrics_per_tag(ground_truth_tags, generated_tags, tag):
    TP = FP = FN = 0
    
    for i in range(len(ground_truth_tags)):
        if ground_truth_tags[i] == tag:
            if generated_tags[i] == tag:
                TP += 1
            else:
                FN += 1
        elif generated_tags[i] == tag:
            FP += 1
    
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0
    
    return precision, recall, f1
# Example usage:

for tag in tags:
    precision, recall, f1 = calculate_metrics_per_tag(ground_truth, BertOutput, tag)
    print(f"Tag: {tag}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1: {f1}")
    print()
```
