```python
import random

def count_sentences_and_get_random_sentences(file_path):
    # Read the content of the file
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()

    # Split the content into sentences
    sentences = content.split('.')

    # Calculate the total number of sentences
    total_sentences = len(sentences)

    # Generate 1000 random indices
    random_indices = random.sample(range(total_sentences), 1000)

    # Get the sentences at the random indices
    random_sentences = [sentences[idx] for idx in random_indices]

    return random_indices, random_sentences

# Replace 'file_path' with the path to your .en file
file_path = '/kaggle/input/dataset/test.en'
random_indices, random_sentences = count_sentences_and_get_random_sentences(file_path)

# Print the random indices and corresponding sentences
# for idx, sentence in zip(random_indices, random_sentences):
#     print(f"Index: {idx}, Sentence: {sentence.strip()}")  # strip() removes leading/trailing whitespaces
```

```python

from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
device = "cuda:0"
```

```python


tokenizer = AutoTokenizer.from_pretrained(
    "facebook/nllb-200-distilled-600M", src_lang="eng_Latn"
)
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
stored_sentence = []
for sentence in random_sentences:
    article = sentence
    inputs = tokenizer(article, return_tensors="pt")

    translated_tokens = model.generate(
        **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["hin_Deva"], max_length=30
    )
    var = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
    print(var)
    stored_sentence.append(var)
print(stored_sentence)
```

```python
def read_sentences_at_indices(file_path, indices):
    sentences = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for index, line in enumerate(file):
            if index in indices:
                sentences.append(line.strip())
    return sentences

# Example usage:
file_path = '/kaggle/input/dataset/test.hi'  # Replace 'your_file.hi' with the path to your .hi file
hindi_sentences = read_sentences_at_indices(file_path, random_indices)
# print(sentences)
```
