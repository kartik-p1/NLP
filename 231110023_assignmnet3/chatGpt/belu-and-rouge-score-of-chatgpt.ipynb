```python
from rouge import Rouge
import nltk
from nltk.translate.bleu_score import corpus_bleu
```

# Bleu and Rouge score of English to hindi convserion using chatGpt

```python
from nltk.translate.bleu_score import corpus_bleu
from rouge import Rouge

# Calculate BLEU score
bleu_score = corpus_bleu(reference, candidate)
print(f"BLEU score: {bleu_score}")
```

```python
rouge = Rouge()
scores = rouge.get_scores(reference, candidate, avg = True)
print(scores)
```
